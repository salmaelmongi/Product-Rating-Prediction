**Q1)Why Data Mining is a misnomer? What is another preferred name?**

The term "data mining" is misleading because it is not about data extraction (mining), but about extracting patterns and knowledge from vast amounts of data. 
It has nothing to do with data extraction, but mainly deals with detecting patterns and noise in datasets.

Data mining, also known as knowledge discovery in databases. 

**Q2)What is the general knowledge discovery process? What is the difference between a data engineer and data scientist/AI engineer?**

KDD has 5 to 7 steps, but many important and authoritative voices on the topic see it as a 5-step process.

1-**Selection**:The data of interest and variables used to evaluate knowledge discovery are determined by acting on a database of compiled data. 

2- **Pre-processing**:This phase is about making the data we work with better and includes the idea of ​​data cleansing. Predictive models for unreliable data are similarly built to predict erroneous, missing, and attribute mismatch data into the future and resolve them from future processes. 

3- **Transformation**:This step focuses on transforming the preprocessed data into a fully usable format. Materials are classified and categorized here, often grouped into one type.  

4- **Data mining**:The data mining state is concerned with sifting through modified data to find interesting patterns. These patterns are graphed, trended, and mapped in ways that are particularly useful for the KDD process. 

5- **Interpretation/evaluation**:The data has been cleaned, translated, split into key criteria and visualized so that people can evaluate the curated results 

**Data engineers** : are concerned with constructing and maintaining data 
infrastructures
 
**data scientists**: are concerned with analysing and interpreting data. Computer scientists, statisticians, and mathematicians are among the most common technical backgrounds for data scientists.

**AI engineers**: create deployable versions of models developed by data scientists and integrate those models into the final product. Data scientists, on the other hand, create machine learning models in IDEs. Optionally, AI developers are also responsible for developing secure web service APIs to deploy models. 


**Q3) In data mining, what is the difference between prediction and categorization?**

Classification is the process of determining the category or class designation to which a new observation belongs, based primarily on current or historical assumptions. Prediction is the process of determining numerical data that are missing or unavailable for new observations, and are primarily based on assumptions about the future.
